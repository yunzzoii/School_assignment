```python
import random
import numpy as np
import math
m = 10000
n = 1000
```

```python
def generate_data(m):
    X = np.empty((1, 0))
    y = np.empty((1, 0))
    for i in range(m):
        degree_value = random.uniform(0, 360)
        cosine_value = math.cos(math.radians(degree_value))
        X = np.append(X, np.array([degree_value]))
        if cosine_value > 0:
            y = np.append(y, np.array([1]))
        else:
            y = np.append(y, np.array([0]))
        X = np.array([X])
        y = np.array([y])
    return X, y
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def logistic_regression_2(X_train, y_train, m, alpha, epoch):
    W1 = np.array([[random.uniform(0, 1)],[random.uniform(0, 1)]]) #(2, 1)
    B1 = np.array([[random.uniform(0, 1)], [random.uniform(0, 1)]])
    W2 = np.array([[random.uniform(0, 1), random.uniform(0, 1)]]) #(1, 2)
    B2 = np.array([[random.uniform(0, 1)]])
    for i in range(epoch): 
        Z1 = np.dot(W1, X_train) + B1 #(2, 1) (1, 10000) -> (2, 10000)
        A1 = sigmoid(Z1) # (2, 10000)
        Z2 = np.dot(W2, A1) + B2 #(1, 2) (2, 10000) -> (1, 10000)
        A2 = sigmoid(Z2)
        
        dZ2 = A2 - y_train # (1, 10000)
        dW2 = np.dot(dZ2, A1.T) / m # (1, 10000) (10000,2) -> (1, 2)
        db2 = np.sum(dZ2, axis=1, keepdims=True) / m 
        W2 = W2 - alpha * dW2
        B2 = B2 - alpha * db2
        
        dZ1 = np.dot(W2.T, dZ2) * sigmoid(Z1) * (1 - sigmoid(Z1)) #(2, 1) (1,10000)  -> (2, 10000)
        dW1 = np.dot(dZ1, X_train.T) / m #(2, 10000) (10000, 1) -> (2, 1)
        db1 = np.sum(dZ1, axis=1, keepdims=True) / m
        W1 = W1 - alpha * dW1
        B1 = B1 - alpha * db1
        if i % 500 == 0:
            print(i, "th result----")
            print("layer 1 parameter : ")
            print("w1_1: ", W1[0][0], "b1_1: ", B1[0][0], "w1_2: ", W1[1][0], "b1_2: ", B1[1][0])
            print("layer 2 parameter : ")
            print("w2_1: ", W2[0][0], "w2_2: ", W2[0][1], "b2: ", B2[0][0])
    return W1, B1, W2, B2
```

```python
def logistic_regression_2_regualization(X_train, y_train, m, alpha, epoch):
    W1 = np.array([[random.uniform(0, 1)],[random.uniform(0, 1)]]) #(2, 1)
    B1 = np.array([[random.uniform(0, 1)], [random.uniform(0, 1)]])
    W2 = np.array([[random.uniform(0, 1), random.uniform(0, 1)]]) #(1, 2)
    B2 = np.array([[random.uniform(0, 1)]])
    X = np.array([[math.radians(x) for x in X_train[0]]]) #(1, 10000)
    for i in range(epoch): 
        Z1 = np.dot(W1, X) + B1 #(2, 1) (1, 10000) -> (2, 10000)
        A1 = sigmoid(Z1) # (2, 10000)
        Z2 = np.dot(W2, A1) + B2 #(1, 2) (2, 10000) -> (1, 10000)
        A2 = sigmoid(Z2)
        
        dZ2 = A2 - y_train # (1, 10000)
        dW2 = np.dot(dZ2, A1.T) / m # (1, 10000) (10000,2) -> (1, 2)
        db2 = np.sum(dZ2, axis=1, keepdims=True) / m 
        W2 = W2 - alpha * dW2
        B2 = B2 - alpha * db2
        
        dZ1 = np.dot(W2.T, dZ2) * sigmoid(Z1) * (1 - sigmoid(Z1)) #(2, 1) (1,10000)  -> (2, 10000)
        dW1 = np.dot(dZ1, X.T) / m #(2, 10000) (10000, 1) -> (2, 1)
        db1 = np.sum(dZ1, axis=1, keepdims=True) / m
        W1 = W1 - alpha * dW1
        B1 = B1 - alpha * db1
        if i % 500 == 0:
            print(i, "th result----")
            print("layer 1 parameter : ")
            print("w1_1: ", W1[0][0], "b1_1: ", B1[0][0], "w1_2: ", W1[1][0], "b1_2: ", B1[1][0])
            print("layer 2 parameter : ")
            print("w2_1: ", W2[0][0], "w2_2: ", W2[0][1], "b2: ", B2[0][0])
    return W1, B1, W2, B2
```

```python
def getCost(y_predict, y_test, n):
    C = 0
    for i, y in enumerate(y_test[0]):
        if (y == 1):
            C += - y * np.log(y_predict[0][i])
        elif (y == 0):
            C += - np.log(1 - y_predict[0][i])
    
    C /= n
    return C
```

```python
def getAcurracy(y_predict, y_test, n):
    y_predict_label = []
    accuracy = 0
    for i, y_pre in enumerate(y_predict[0]):
        if y_pre > 0.5:
            y_predict_label.append(1)
        else:
            y_predict_label.append(0)
    for i, y in enumerate(y_test[0]):
        if y == y_predict_label[i]:
            accuracy += 1
    accuracy = accuracy / n * 100
    return accuracy
    
```

```python
def predict(X_test, y_test, n, W1, B1, W2, B2):
    Z1 = np.dot(W1, X_test) + B1 
    A1 = sigmoid(Z1) # (1000, 1)
    Z2 = np.dot(W2, A1) + B2
    A2 = sigmoid(Z2)
    C = getCost(A2, y_test, n)
    print("Cost: ", C)
    print("Accuracy: ", getAcurracy(A2, y_test, n) )
```

```python
def predict_regulization(X_test, y_test, n, W1, B1, W2, B2):
    X = np.array([[math.radians(x) for x in X_test[0]]]) #(1, 10000)
    Z1 = np.dot(W1, X) + B1 
    A1 = sigmoid(Z1) # (1000, 1)
    Z2 = np.dot(W2, A1) + B2
    A2 = sigmoid(Z2)
    C = getCost(A2, y_test, n)
    print("Cost: ", C)
    print("Accuracy: ", getAcurracy(A2, y_test, n) )
```
