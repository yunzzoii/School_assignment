```python
import random
import numpy as np
import math
m = 10000
n = 1000
```

```python
def generate_data(m):
    X = np.empty((0, 1))
    y = np.empty((0, 1))
    for i in range(m):
        degree_value = random.uniform(0, 360)
        cosine_value = math.cos(math.radians(degree_value))
        X = np.append(X, np.array([[degree_value]]), axis=0)
        if cosine_value > 0:
            y = np.append(y, np.array([[1]]), axis=0)
        else:
            y = np.append(y, np.array([[0]]), axis=0)
    return X, y
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def logistic_regression_2(X_train, y_train, m, alpha, epoch):
    w1 =  w2 = b1 = b2 = 1
    W1 = np.array([[w1]])
    W2 = np.array([[w2]])
    for i in range(epoch):
        #X = np.array([[math.radians(x[0]) for x in X_train]])
        Z1 = np.dot(X_train, W1.T) + b1 #(10000, 1) (1, 1) -> (10000, 1) 
        A1 = sigmoid(Z1)
        Z2 = np.dot(A1, W2.T) + b2 #(10000, 1) (1, 1) -> (10000, 1) 
        A2 = sigmoid(Z2)
        
        dZ2 = A2 - y_train # (10000, 1)
        dW2 = np.dot(dZ2.T, A1) / m # (1, 10000) (10000, 1) -> (1, 1)
        db2 = np.sum(dZ2) / m
        W2 = W2 - alpha * dW2
        b2 = b2 - alpha * db2
        
        dZ1 = np.dot(dZ2, W2) * sigmoid(Z1) * (1 - sigmoid(Z1)) #(10000, 1) (1, 1)
        dW1 = np.dot(dZ1.T, X_train) / m #(1, 10000) (10000, 1)
        db1 = np.sum(dZ1) / m
        W1 = W1 - alpha * dW1
        b1 = b1 - alpha * db1
        if i % 500 == 0:
            print("w1: ", W1[0][0], "b1: ", b1)
            print("w2: ", W2[0][0], "b2: ", b2)
    return W1, b1, W2, b2
```

```python
def logistic_regression_2_regualization(X_train, y_train, m, alpha, epoch):
    w1 =  w2 = b1 = b2 = 1
    W1 = np.array([[w1]])
    W2 = np.array([[w2]])
    X = np.array([[math.radians(x[0]) for x in X_train]])
    for i in range(epoch): 
        Z1 = np.dot(X.T, W1) + b1 #(10000, 1) (1, 1) -> (10000, 1) 
        A1 = sigmoid(Z1)
        Z2 = np.dot(A1, W2.T) + b2 #(10000, 1) (1, 1) -> (10000, 1) 
        A2 = sigmoid(Z2)
        
        dZ2 = A2 - y_train # (10000, 1)
        dW2 = np.dot(dZ2.T, A1) / m # (1, 10000) (10000, 1) -> (1, 1)
        db2 = np.sum(dZ2) / m
        W2 = W2 - alpha * dW2
        b2 = b2 - alpha * db2
        
        dZ1 = np.dot(dZ2, W2) * sigmoid(Z1) * (1 - sigmoid(Z1)) #(10000, 1) (1, 1)
        dW1 = np.dot(X, dZ1) / m #(1, 10000) (10000, 1)
        db1 = np.sum(dZ1) / m
        W1 = W1 - alpha * dW1
        b1 = b1 - alpha * db1
        if i % 500 == 0:
            print(i, "th result")
            print("w1: ", W1[0][0], "b1: ", b1)
            print("w2: ", W2[0][0], "b2: ", b2)
    return W1, b1, W2, b2
```

```python
def getCost(y_predict, y_test, n):
    C = 0
    for i, y in enumerate(y_test):
        if (y == 1):
            C += - y * np.log(y_predict[i])
        elif (y == 0):
            C += - np.log(1 - y_predict[i])
    
    C /= n
    return C
```

```python
def getAcurracy(y_predict, y_test, n):
    y_predict_label = []
    accuracy = 0
    for i, y_pre in enumerate(y_predict):
        if y_pre > 0.5:
            y_predict_label.append(1)
        else:
            y_predict_label.append(0)
    for i, y in enumerate(y_test):
        if y == y_predict_label[i]:
            accuracy += 1
    accuracy = accuracy / n * 100
    return accuracy
    
```

```python
def predict(X_test, y_test, n, W1, b1, W2, b2):
    # X = np.array([[math.radians(x[0]) for x in X_test]])
    Z1 = np.dot(X_test, W1.T) + b1
    A1 = sigmoid(Z1) # (1000, 1)
    Z2 = np.dot(A1, W2.T) + b2
    A2 = sigmoid(Z2)
    C = getCost(A2, y_test, n)
    print("Cost: ", C[0])
    print("Accuracy: ", getAcurracy(A2, y_test, n) )
```

```python
def predict_regulization(X_test, y_test, n, W1, b1, W2, b2):
    X = np.array([[math.radians(x[0]) for x in X_test]])
    Z1 = np.dot(X.T, W1) + b1
    A1 = sigmoid(Z1) # (1000, 1)
    Z2 = np.dot(A1, W2.T) + b2
    A2 = sigmoid(Z2)
    C = getCost(A2, y_test, n)
    print("Cost: ", C[0])
    print("Accuracy: ", getAcurracy(A2, y_test, n) )
```

```python
m = 10000
n = 1000
X_train, y_train = generate_data(m)
X_test, y_test = generate_data(n)
W1, b1, W2, b2 = logistic_regression_2(X_train, y_train, m, alpha=0.9, epoch=5000)
print("--------------------")
print("w1: ", W1[0][0], "b1: ", b1)
print("w2: ", W2[0][0], "b2: ", b2)
print("Train result")
predict(X_train, y_train, m, W1, b1, W2, b2)
print("Test result")
predict(X_test, y_test, n, W1, b1, W2, b2)
```
